```python
import sympy as sym


# Not necessary but gives nice-looking latex output
# More info at: http://docs.sympy.org/latest/tutorial/printing.html
sym.init_printing()

t = sym.symbols('t')
m1, m2, m3 = sym.symbols('m_1 m_2 m_3', positive=True)
j1= sym.Function('phi')(t)
j2= sym.Function('psi')(t)
j3= sym.Function('chi')(t)

x = sym.Function('x')(t)
y = sym.Function('y')(t)
z = sym.Function('z')(t)

vec_new = sym.Matrix([j1, j2, j3])
vec_old = sym.Matrix([x, y, z])

coefs = sym.Matrix([2*m1, 2*m2, 2*m3])

M = m1 + m2 + m3
matrix = sym.Matrix([[1, -1, 0], [m1/M, m2/M, -1], [m1/M, m2/M, m3/M]])
```

```python
inv = sym.simplify(matrix.inv()).applyfunc(sym.factor)
inv
```

```python
vec_old = inv @ vec_new
vec_old = vec_old.diff(t)
vec_old = vec_old.applyfunc(lambda x:x**2)
for i in range(3):
    vec_old[i] *= 1/coefs[i]
vec_old.applyfunc(sym.factor)
#vec_old
```
